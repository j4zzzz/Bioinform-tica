# Ejercicio 1 y 2

```python
secuencias_ejercicio_1 = [
    "ATCGTACGATGACCTGATCG",
    "GGTATACGATGACGTTACCA",
    "TTTCTACGATGACCATAGGT",
    "AACGTACGATGACGGGTTAA",
    "CGGATACGATGACTTCCGTA",
    "TACCTACGATGACAGGTACA",
    "GACTTACGATGACCGATAGC",
    "TCGATACGATGACTGGCAAT",
    "AGGCTACGATGACATTCGGA",
    "CCTATACGATGACGGAATTC"
]

def obtener_k_mers_frecuentes(secuencias, k):
    conteo_k_mers = {}
    for secuencia in secuencias:
        k_mers_vistos = set()
        for i in range(len(secuencia) - k + 1):
            k_mer = secuencia[i:i+k]
            k_mers_vistos.add(k_mer)
        for k_mer in k_mers_vistos:
            conteo_k_mers[k_mer] = conteo_k_mers.get(k_mer, 0) + 1

    max_frecuencia = max(conteo_k_mers.values())
    k_mers_mas_frecuentes = {k_mer: frec for k_mer, frec in conteo_k_mers.items() if frec == max_frecuencia}
    return k_mers_mas_frecuentes

k_mers_7 = obtener_k_mers_frecuentes(secuencias_ejercicio_1, 7)
k_mers_8 = obtener_k_mers_frecuentes(secuencias_ejercicio_1, 8)
k_mers_9 = obtener_k_mers_frecuentes(secuencias_ejercicio_1, 9)

print("Mejores k-mers k=7:", k_mers_7)
print("Mejores k-mers k=8:", k_mers_8)
print("Mejores k-mers k=9:", k_mers_9)
```

# Ejercicio 3

```python
mejor_k_mer_9 = list(k_mers_9.keys())[0]

posiciones_ocurrencias = []
for i, secuencia in enumerate(secuencias_ejercicio_1):
    posicion = secuencia.find(mejor_k_mer_9)
    posiciones_ocurrencias.append((i, posicion))

print("Posiciones del k-mer candidato en cada secuencia:", posiciones_ocurrencias)
```

# Ejercicio 4 Y 5

```python
alineamiento_multiple = []
longitud_motif = len(mejor_k_mer_9)

for secuencia in secuencias_ejercicio_1:
    posicion = secuencia.find(mejor_k_mer_9)
    if posicion != -1:
        region_extraida = secuencia[posicion:posicion+longitud_motif]
        alineamiento_multiple.append(region_extraida)

for secuencia_alineada in alineamiento_multiple:
    print(secuencia_alineada)
```

# Ejercicio 6

```python
nucleotidos = ['A', 'C', 'G', 'T']
matriz_frecuencias = {n: [0] * longitud_motif for n in nucleotidos}

for secuencia in alineamiento_multiple:
    for posicion, nucleotido in enumerate(secuencia):
        matriz_frecuencias[nucleotido][posicion] += 1

for nucleotido in nucleotidos:
    print(f"{nucleotido}: {matriz_frecuencias[nucleotido]}")
```

# Ejercicio 7

```python
secuencia_consenso = ""

for posicion in range(longitud_motif):
    max_frecuencia = -1
    nucleotido_frecuente = ""
    for nucleotido in nucleotidos:
        frecuencia_actual = matriz_frecuencias[nucleotido][posicion]
        if frecuencia_actual > max_frecuencia:
            max_frecuencia = frecuencia_actual
            nucleotido_frecuente = nucleotido
    secuencia_consenso += nucleotido_frecuente

print("Secuencia consenso:", secuencia_consenso)
```

# Ejercicio 8

```python
total_secuencias = len(alineamiento_multiple)
porcentajes_conservacion = []
posiciones_totalmente_conservadas = []
posiciones_variables = []

for posicion in range(longitud_motif):
    nucleotido_consenso = secuencia_consenso[posicion]
    frecuencia_consenso = matriz_frecuencias[nucleotido_consenso][posicion]
    porcentaje = (frecuencia_consenso / total_secuencias) * 100
    porcentajes_conservacion.append(porcentaje)

    if porcentaje == 100.0:
        posiciones_totalmente_conservadas.append(posicion)
    else:
        posiciones_variables.append(posicion)

porcentaje_global = sum(porcentajes_conservacion) / longitud_motif

print("Porcentajes por posición:", porcentajes_conservacion)
print("Posiciones 100% conservadas:", posiciones_totalmente_conservadas)
print("Posiciones variables:", posiciones_variables)
print("Conservación global (%):", porcentaje_global)
```
